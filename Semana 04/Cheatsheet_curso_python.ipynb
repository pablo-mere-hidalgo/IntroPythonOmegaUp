# Estructuras de flujo iterativas

Aparte de tener condicionales, tambi칠n tenemos este tipo de estructuras de flujo, que nos permiten ejecutar un mismo c칩digo de manera repetida mientras se cumpla una condici칩n. Existen dos estructuras c칤clicas en python:
* `for`
* `while`



## Ciclo for

Este ciclo nos permite iterar operaciones. Para entenderlo mejor, vamos a hacer algunos ejemplos. Digamos que tenemos una lista de personas, y queremos saludar a todos los de la lista, pero es cansado escribir todos los prints, entonces lo podemos hacer as칤

```python
lista_pokemon=['Sceptile','Mightyena','Flygon','Salamance','Charizard','Mewtwo']
for pokemon in lista_pokemon:
    print(f"Hola {pokemon}!")
```

Tambi칠n podemos hacerlo con numeros, digamos que queremos imprimir solo los pares del 50 al 100

```python
i=0
for numero in range(i,101):
    if(numero%3==0):
        print(f"El numero {numero} es multiplo de 3")
```

## Ciclo while

Este ciclo siempre se ejecutar치 mientras se cumpla su condici칩n asociada, veamoslo con un ejemplo.

Digamos que quiero comprar dulces, pero para eso necesito dinero, puesto que cada dulce cuesta 5 pesos. Pero digamos que solo comprar칠 dulces mientras tenga m치s de 50 pesos. entonces:

```python
dinero= 200
i=0
while(dinero>50):
    dinero=dinero-12
    print(f"Te compraste un arizona de frambuesa y te quedan {dinero} pesos, chale")
    i+=1
print(f"Al final te compraste {i} arizonas y gastaste en total {i*12} pesos")
```
